# Nettoyage

### Medals

```python
# Remplacement des lignes nulles

df_medals_cleaned = df_medals.withColumn("participant_title", 
                                         when(col("participant_title").isNull(), col("country_name")).otherwise(col("participant_title")))

df_medals_cleaned = df_medals_cleaned.withColumn("athlete_url", 
                                         when(col("athlete_url").isNull(), "").otherwise(col("athlete_url")))

df_medals_cleaned = df_medals_cleaned.withColumn("athlete_full_name", 
                                         when(col("athlete_full_name").isNull(), "").otherwise(col("athlete_full_name")))
```

```python
df_medals_cleaned.show(1000, truncate=False)
```

### Hosts

```python
df_hosts
```

```python
# Suppression des entrées concernant les jeux d'hiver
df_host_clean = df_hosts.filter(df_hosts.game_season == "Summer")
df_host_clean.show(truncate=False)
```

```python
# On renomme cette colonne pour faire la fusion des datasets plus tard (histoire que le nom soit en commun avec les deux datasets)
df_host_clean = df_host_clean.withColumnRenamed('game_slug', 'slug_game')
```

### Athletes

```python
df_athletes_clean = df_athletes.drop('bio','athlete_year_birth','athlete_medals')
df_athletes_clean.show()
```

### Results

```python
df_results_clean = df_results.drop('value_unit','value_type','athlete_url','athlete_full_name')
# df_results_clean = df_results_clean.fillna(0)

# Remplacement des "NaN" dans les types de médailles par "AUCUNE"
df_results_clean = df_results_clean.withColumn("medal_type", when(df_results_clean["medal_type"] == "NaN", "AUCUNE").otherwise(df_results_clean["medal_type"]))
df_results_clean.show(truncate=False)
```

# Fusion des datasets

### Première fusion - A partir de la table Medals

#### Hosts + Medals

```python
df_merged_medals = df_medals_cleaned.join(df_host_clean, on="slug_game", how="inner")
```

#### + Athletes

```python
df_merged_medals = df_merged_medals.join(df_athletes_clean, on="athlete_full_name", how="inner")
```

```python
df_merged_medals
```

```python
df_merged_medals.show(truncate=False)
```

### Seconde fusion - A partir de la table Results

#### Results + Host

```python
df_merged_results = df_results_clean.join(df_host_clean, on="slug_game", how="inner")
```

```python
df_merged_results.show()
```
